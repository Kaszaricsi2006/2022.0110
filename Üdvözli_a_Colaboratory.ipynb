## Adattudomány
</div>

A Colab segítségével a népszerű Python-függvénytárak minden lehetőségével elemezhet és vizualizálhat adatokat. Az alábbi kódcella a <strong>numpy</strong> használatával generál véletlenszerű adatokat, és a <strong>matplotlib</strong> segítségével vizualizálja őket. A kód szerkesztéséhez kattintson a cellára, és végezze el a kívánt módosításokat.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

def plot_function(x_range):
    x = np.linspace(x_range[0], x_range[1], 100)
    y = x ** 2

    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Function Plot: y = x ** 2')
    plt.grid(True)
    plt.show()

# Függvény ábrázolása
plot_function([-5, 5])
```
